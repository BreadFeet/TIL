# Stack  

```python
#   
#     
def isStackFull():
    global SIZE, stack, top
    if top >= SIZE - 1:    # == SIZE-1      
        return True
    else:
        return False

    
# Push 
def push(data):
    global SIZE, stack, top
    if isStackFull():
        print('Stack is FULL--------------')
        return
    top += 1
    stack[top] = data

    
#     
def isStackEmpty():
    global SIZE, stack, top
    if top <= -1:
        return True
    else:
        return False
   

# Pop 
def pop():
    global SIZE, stack, top
    if isStackEmpty():
        print('Stack is EMPTY---------------')
    data = stack[top]
    stack[top] = None
    top -= 1
    return data
    

# Pick (  )
def pick():
    global SIZE, stack, top
    if isStackEmpty():
        print('Stack is EMPTY----------------')
        return None
    return stack[top]



#   
SIZE = 5
stack = [None for _ in range(5)]
top = -1


#    
# isStackFull   --------------------------------
# stack = ['', '', '', '', '']
# top = 4
# isStackFull()

# Push   ---------------------------------------
# stack = ['', '', '', '', None]
# top = 3
# print(stack)
# push('')
# print(stack)
# push('')

# Pop   ----------------------------------------
# stack = ['', None, None, None, None]
# top = 0
# print(stack)
# retData = pop()
# print('Poped data:', retData)
# print(stack)
# retData = pop()

# Pick   ---------------------------------------
# stack = ['', '', '', None, None]
# top = 2
# print(stack)
# retData = pick()
# print('Picked data:', retData)
# print(stack)
```

```python
#  
if __name__ == '__main__':
    select = input('I: , E: , C: , Q: ')
    
    while select != 'X':
        if select == 'I':
            data = input('Input data-----------------')
            push(data)
            print('Current stack:', stack)
        
        elif select == 'E':
            retData = pop()
            print('Poped data:', retData)
            print('Current stack:', stack)
        
        elif select == 'C':
            retData = pick()
            print('Picked data:', retData)
            print('Current stack:', stack)
           
        else:
            print('  .')
        
        select = input('I: , E: , C: , Q: ')
    print(' --------------------------')
```
