# Stack

```python
SIZE = 5
stack  = [None for _ in range(SIZE)]   # stack 비어있는 상태
top = -1    # 높이 초기 설정
```

## Code for 'Push' Behavior

```python
top += 1
stack[top] = '커피'
stack
```

```python
top += 1
stack[top] = '녹차'
stack
```

```python
top += 1
stack[top] = '꿀물'
stack
```

## Code for 'Pop' Behavior

```python
data = stack[top]    # 현재 제일 위 음료
stack[top] = None    # 제일 위 음료를 뺀다
top -= 1             # 높이 한칸 낮춤
print('빼낸 음료 -->', data)
```

```python
data = stack[top]
stack[top] = None
top -= 1
print('빼낸 음료 -->', data)
```

# Stack의 일반적 구현

```python
# 함수 선언 부분
# 스택이 꽉 찾는지 확인하는 함수
def isStackFull():
    global SIZE, stack, top
    if top >= SIZE - 1:    # == SIZE-1 도 맞는데 일반적으로 크거나 같다고 쓴다
        return True
    else:
        return False

    
# Push 함수
def push(data):
    global SIZE, stack, top
    if isStackFull():
        print('Stack is FULL--------------')
        return
    top += 1
    stack[top] = data

    
# 스택이 이미 비었는지 확인하는 함수
def isStackEmpty():
    global SIZE, stack, top
    if top <= -1:
        return True
    else:
        return False
   

# Pop 함수
def pop():
    global SIZE, stack, top
    if isStackEmpty():
        print('Stack is EMPTY---------------')
    data = stack[top]
    stack[top] = None
    top -= 1
    return data
    

# Pick 함수(데이터 확인하는 함수)
def pick():
    global SIZE, stack, top
    if isStackEmpty():
        print('Stack is EMPTY----------------')
        return None
    return stack[top]



# 전역 변수 부분
SIZE = 5
stack = [None for _ in range(5)]
top = -1


# 메인 함수 부분 
# isStackFull 함수 작동 확인--------------------------------
# stack = ['커피', '녹차', '꿀물', '콜라', '환타']
# top = 4
# isStackFull()

# Push 함수 작동 확인---------------------------------------
# stack = ['커피', '녹차', '꿀물', '콜라', None]
# top = 3
# print(stack)
# push('게토레이')
# print(stack)
# push('환타')

# Pop 함수 작동 확인----------------------------------------
# stack = ['커피', None, None, None, None]
# top = 0
# print(stack)
# retData = pop()
# print('Poped data:', retData)
# print(stack)
# retData = pop()

# Pick 함수 작동 확인---------------------------------------
# stack = ['커피', '녹차', '꿀물', None, None]
# top = 2
# print(stack)
# retData = pick()
# print('Picked data:', retData)
# print(stack)
```

```python
# 프로그램 완성
if __name__ == '__main__':
    select = input('I: 삽입, E: 추출, C: 확인, Q: 종료')
    
    while select != 'X':
        if select == 'I':
            data = input('Input data-----------------')
            push(data)
            print('Current stack:', stack)
        
        elif select == 'E':
            retData = pop()
            print('Poped data:', retData)
            print('Current stack:', stack)
        
        elif select == 'C':
            retData = pick()
            print('Picked data:', retData)
            print('Current stack:', stack)
           
        else:
            print('올바른 값을 입력하세요.')
        
        select = input('I: 삽입, E: 추출, C: 확인, Q: 종료')
    print('프로그램을 종료합니다--------------------------')
```
