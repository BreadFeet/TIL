```python
# Factorial
def factorial(num):
    if num <= 1:
        print('1 ')
        return 1
    print(f'factorial({num-1}) ')
    fac = factorial(num-1)
    
    print(f'{num} * factorial({num-1}) ')
    return num * fac

factorial(5)
```

```python
# Countdown
def countDown(n):
    if n == 0:
        print('Launch!')
    else:
        print(n)
        countDown(n-1)
        print(n-1)          #     

countDown(5)
```

# Sorting
## Selection Sort

```python
#   
def findMinIdx(array):
    minIdx = 0
    for i in range(1, len(array)):
        if array[minIdx] > array[i]:
            minIdx = i
    return minIdx

#   
testArray = [55, 88, 33, 77]
minPos = findMinIdx(testArray)
print(':', testArray[minPos])
```

```python
import random
before = [random.randint(100, 999) for _ in range(8)]   #   array
after = []     #   array

print(' :', before)

for i in range(len(before)):    #      after  
    minPos = findMinIdx(before)
    after.append(before[minPos])
    del before[minPos]

print(' :', after)
```

## Impoved Selection Sort


```python
# 
def selectionSort(array):
    n = len(array)    #      
    for i in range(0, n-1):
        minIdx = i
        for j in range(i+1, n):   #  
            if array[minIdx] > array[j]:
                minIdx = j
        array[i], array[minIdx] = array[minIdx], array[i]
    return array


# 
import random
dataArray = [random.randint(100, 999) for _ in range(8)]

# 
print(' :', dataArray)
dataArray = selectionSort(dataArray)
print(' :', dataArray)
```

# Search
## Binary Search


```python
#  
def binarySearch(array, findData):
    pos = -1        # pos:       
    start = 0
    end = len(array) - 1
    
    while start <= end:
        mid = (start + end) // 2      #  index
        if array[mid] == findData:     #    
            return mid
        elif findData > array[mid]:
            start = mid + 1
        else:        # findData < array[mid] 
            end = mid - 1
   
    # mid    start >= end   
    return pos
            
    
#  
dataArray = [50, 60, 105, 120, 150, 160, 162, 168, 177, 182]
findData = 162


#  
print(' :', dataArray)
position = binarySearch(dataArray, findData)
if position == -1:
    print(findData, ':   .')
else:    # mid   
    print(findData, ':  ', position, ' .')
```
