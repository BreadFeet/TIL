```python
def openBox():
    global count
    if count == 0:
        print('Put my precious ring in the', count, 'box.')
        print('Close the', count, 'box.')
        return
    print('Open the', count, 'box.')
    count -= 1
    openBox()
    count += 1
    print('Close the', count, 'box.')

    
count = 10
openBox()
```

```python
# Factorial
def factorial(num):
    if num <= 1:
        print('1 반환')
        return 1
    print(f'factorial({num-1}) 호출')
    fac = factorial(num-1)
    
    print(f'{num} * factorial({num-1}) 반환')
    return num * fac

factorial(5)
```

```python
# Countdown
def countDown(n):
    if n == 0:
        print('Launch!')
    else:
        print(n)
        countDown(n-1)
        print(n-1)          # 거꾸로 반환되는 것 확인하려고 추가

countDown(5)
```

# Sorting
## Selection Sort

```python
# 함수 선언 부분
def findMinIdx(array):
    minIdx = 0
    for i in range(1, len(array)):
        if array[minIdx] > array[i]:
            minIdx = i
    return minIdx

# 메인 함수 코드
testArray = [55, 88, 33, 77]
minPos = findMinIdx(testArray)
print('최소값:', testArray[minPos])
```

```python
import random
before = [random.randint(100, 999) for _ in range(8)]   # 정렬 전 array
after = []     # 정렬 후 array

print('정렬 전:', before)

for i in range(len(before)):    # 남은 것 중 최소값을 찾아서 after에 넣고 삭제
    minPos = findMinIdx(before)
    after.append(before[minPos])
    del before[minPos]

print('정렬 후:', after)
```

## Impoved Selection Sort


```python
# 함수
def selectionSort(array):
    n = len(array)    # 몇번 사이클을 반복해야 하는지 확인하기 위해서
    for i in range(0, n-1):
        minIdx = i
        for j in range(i+1, n):   # 비교할 범위
            if array[minIdx] > array[j]:
                minIdx = j
        array[i], array[minIdx] = array[minIdx], array[i]
    return array


# 변수
import random
dataArray = [random.randint(100, 999) for _ in range(8)]

# 메인
print('정렬 전:', dataArray)
dataArray = selectionSort(dataArray)
print('정렬 후:', dataArray)
```

# Search
## Binary Search


```python
# 함수 정의
def binarySearch(array, findData):
    pos = -1        # pos: 값을 못 찾은 경우 반환하기 위한 값
    start = 0
    end = len(array) - 1
    
    while start <= end:
        mid = (start + end) // 2      # 중앙값 index
        if array[mid] == findData:     # 중앙값이 찾고자 하는 값
            return mid
        elif findData > array[mid]:
            start = mid + 1
        else:        # findData < array[mid]인 경우
            end = mid - 1
   
    # mid를 아직 찾지 못했는데 start >= end로 엇갈리면 반복문 탈출
    return pos
            
    
# 변수 지정
dataArray = [50, 60, 105, 120, 150, 160, 162, 168, 177, 182]
findData = 162


# 메인 코드
print('정렬 전:', dataArray)
position = binarySearch(dataArray, findData)
if position == -1:
    print(findData, ': 이 값은 없습니다.')
else:    # mid 값이 반환된 경우
    print(findData, ': 이 값은', position, '위치에 있습니다.')
```
