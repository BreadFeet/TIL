```python
#  ( )
def add_data(friend):
    kakao.append(None)    # list append  
    kLen = len(kakao)
    kakao[kLen-1] = friend     #    index 
    
def insert_data(position, friend):
    kakao.append(None)     #  
    kLen = len(kakao)
    for i in range(kLen-1, position, -1):  #  position  -1 
        kakao[i] = kakao[i-1]    #    
        kakao[i-1] = None        #    
    #  position  
    kakao[position] = friend
    
def delete_data(position):
    kLen = len(kakao)
    kakao[position] = None
    for i in range(position+1, kLen, 1):
        kakao[i-1] = kakao[i]
        kakao[i] = None
    #     -  
    del kakao[kLen-1]


#   
kakao = []
select = -1   # 1: , 2: , 3:, 4: 


#   
if __name__ == '__main__':
    while select != 4:    #     
        select = int(input('1:, 2:, 3:, 4:  .'))
        
        if select == 1:
            friend = input('  ')
            add_data(friend)
            print(kakao)
        
        elif select == 2:
            position = int(input(' '))
            friend = input('  ')
            insert_data(position, friend)
            print(kakao)
            
        elif select == 3:
            position = int(input(' '))
            delete_data(position)
            print(kakao)
            
        elif select == 4:
            print(kakao)
            #break
            
        else:
            print(' (1~4) ')   # 1~4   .   int  
            #continue
```
