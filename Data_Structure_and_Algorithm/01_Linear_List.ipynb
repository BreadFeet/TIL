```python
kakao = ['다현', '정연', '쯔위', '사나', '지효']
```

```python
def insert_data(position, friend):
    kakao.append(None)     # 빈자리 만듦
    kLen = len(kakao)
    for i in range(kLen-1, position, -1):  # 마지막부터 position 뒤까지 -1 반복
        kakao[i] = kakao[i-1]    # 앞사람을 한칸 뒤로 옮김
        kakao[i-1] = None        # 앞사람 자리 빈칸으로 만듦
    # 마지막에 position 위치가 빔
    kakao[position] = friend
```

```python
insert_data(2, '솔라')
kakao
```

```python
kakao = ['다현', '정연', '쯔위', '사나', '지효']

def delete_data(position):
    kLen = len(kakao)
    kakao[position] = None
    for i in range(position+1, kLen, 1):
        kakao[i-1] = kakao[i]
        kakao[i] = None
    # 마지막 칸이 비어있는 상태 - 빈칸 삭제
    del kakao[kLen-1]
```

```python
# 정연의 1 위치를 삭제
delete_data(1)
kakao
```

## 선형 리스트 일반적 구현 완성

```python
# 함수 선언부(클래스 선언)
def add_data(friend):
    kakao.append(None)    # list의 append는 순서대로 추가
    kLen = len(kakao)
    kakao[kLen-1] = friend     # 입력된 친구가 들어가는 index 지정
    
def insert_data(position, friend):
    kakao.append(None)     # 빈자리 만듦
    kLen = len(kakao)
    for i in range(kLen-1, position, -1):  # 마지막부터 position 뒤까지 -1 반복
        kakao[i] = kakao[i-1]    # 앞사람을 한칸 뒤로 옮김
        kakao[i-1] = None        # 앞사람 자리 빈칸으로 만듦
    # 마지막에 position 위치가 빔
    kakao[position] = friend
    
def delete_data(position):
    kLen = len(kakao)
    kakao[position] = None
    for i in range(position+1, kLen, 1):
        kakao[i-1] = kakao[i]
        kakao[i] = None
    # 마지막 칸이 비어있는 상태 - 빈칸 삭제
    del kakao[kLen-1]


# 전역 변수 부분
kakao = []
select = -1   # 1:뒤에 추가, 2:안에 삽입, 3:삭제, 4: 종료


# 메인 코드 부분
if __name__ == '__main__':
    while select != 4:    # 종료 요청이 아닌 동안 반복
        select = int(input('1:추가, 2:삽입, 3:삭제, 4:종료 중 선택하세요.'))
        
        if select == 1:
            friend = input('추가할 친구 이름')
            add_data(friend)
            print(kakao)
        
        elif select == 2:
            position = int(input('삽입 위치'))
            friend = input('삽입할 친구 이름')
            insert_data(position, friend)
            print(kakao)
            
        elif select == 3:
            position = int(input('삭제할 위치'))
            delete_data(position)
            print(kakao)
            
        elif select == 4:
            print(kakao)
            #break
            
        else:
            print('올바른 숫자(1~4)를 입력하세요')   # 1~4 이외 숫자이면 작동. 문자열이 들어오면 int가 안되서 오류남
            #continue
```
